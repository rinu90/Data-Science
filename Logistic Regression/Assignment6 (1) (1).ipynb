```python
y = data.iloc[:,18:19]
```

```python
df1=data.copy()
```

```python
df1 = df1.drop('y_yes', axis=1)
```

```python
x = df1.iloc[:,0:42]
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x,y)
```

```python
y_pred = classifier.predict(x)
```

```python
y1 = data.iloc[:,18]
```

```python
y1
```

```python
y_pred_df = pd.DataFrame({'actual':y1,'predicted':y_pred})
```

```python
from sklearn.metrics import confusion_matrix
```

```python
ConfusionM = confusion_matrix(y1,y_pred)
```

```python
print(ConfusionM)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y1,y_pred))
```

```python
from sklearn.metrics import roc_curve
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y2= y1.to_numpy()
```

```python
frp,trp,thresholds= roc_curve(y2,classifier.predict_proba(x)[:,1])
```

```python
auc = roc_auc_score(y2,y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(frp,trp,color = 'red', label ='logit mode(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False positive rate')
plt.ylabel('true positive rate')
```
