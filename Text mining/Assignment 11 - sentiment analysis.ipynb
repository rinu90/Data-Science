```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.amazon.in/Samsung-Midnight-Storage-6000mAh-Battery/product-reviews/B0B4F38D6K/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1'
```

```python
r = requests.get(url,params={'url': url, 'wait': 2})
r
```

```python
soup=BeautifulSoup(r.text, "lxml")
```

```python
a=soup.find_all("span", class_="a-size-base")
P=[]

for i in a:
    a=i.text
    P.append(a)
print(P)  
```

```python
import pandas as pd
```

```python
import numpy as np
rev = np.array(P)
```

```python
review=[]
for i in rev:
    review.append(i.strip())
        
        
    
```

```python
review
```

```python
import string
# Remove punctuation from each string in the list
no_punc_reviews = [text.translate(str.maketrans('', '', string.punctuation)) for text in review]

# Display only the cleaned text
for i in no_punc_reviews:
    cleaned_text=cleaned_text+i
```
