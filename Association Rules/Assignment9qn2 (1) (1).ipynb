```python
!pip install mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(data, min_support=0.1, use_colnames =True)
```

```python
frequent_itemsets
```

```python
#for different min length
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets[ (frequent_itemsets['length'] == 2)]
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 3)]
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 4)]
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold=.7)
```

```python
rules
```

```python
#different values of support and confidence

rules[(rules['support'] >.1) &(rules['confidence']>=.4)]
```

```python
rules[(rules['support'] >.2) &(rules['confidence']>=.6)]
```
