```python
import chardet
with open("/content/book (1).csv", "rb") as f:
  encoding = chardet.detect(f.read())["encoding"]
```

```python
data = pd.read_csv("/content/book (1).csv",encoding=encoding)
```

```python
reshaped_df = data.pivot_table(index='User.ID', columns='Book.Title', values='Book.Rating')
reshaped_df
```

```python
reshaped_df.fillna(0,inplace=True)
```

```python
reshaped_df.index
```

```python
reshaped_df.head()
```

```python
from sklearn.metrics import pairwise_distances
```

```python
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(reshaped_df.values, metric = 'cosine')
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim,0)
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
user_sim_df.index = user_sim_df.columns = reshaped_df.index
```

```python
user_sim_df
```

```python
data[(data['User.ID']==8)|(data['User.ID']==9)]
```

```python
user1 = data[(data['User.ID']==8)]
```

```python
user2 = data[(data['User.ID']==9)]
```

```python
user1
```

```python
user2
```

```python
pd.merge(user1, user2, on='Book.Rating', how= 'outer')
```

```python
# Find similar users for a given user
user_id = 8
# Retrieve cosine similarity scores for the user
similar_users = user_sim_df[user_id].sort_values(ascending=False)
```

```python
# Display the top similar users (excluding the user itself)
top_similar_users = similar_users.drop(user_id).head(5)
print("Top similar users for user", user_id, ":\n", top_similar_users)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.histplot(data['Book.Rating'], bins=10, kde=True)
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

#the distribution of ratings given by users.
```

```python
plt.figure(figsize=(8, 8))
rating_distribution = data['Book.Rating'].value_counts()
plt.pie(rating_distribution, labels=rating_distribution.index, autopct='%1.1f%%', colors=['lightgreen', 'lightblue', 'lightcoral', 'lightskyblue', 'lightpink'])
plt.title('Distribution of Book Ratings')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
rating_counts = data['Book.Title'].value_counts().head(10)
rating_counts.plot(kind='bar', color='salmon')
plt.title('Top 10 Most Rated Books')
plt.xlabel('Book Title')
plt.ylabel('Rating Count')
plt.xticks(rotation=45)
plt.show()
```
